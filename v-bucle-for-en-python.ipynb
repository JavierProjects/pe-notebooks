# 1. Instroducción 
[[Contenido]](#Contenido)

En esta lección aprenderemos cómo usar un bucle `for` para repetir un proceso y analizar un conjunto de datos, y cómo leer un conjunto de datos grande como una lista de listas.

En ejercicios anteriores cuando trabajábamos con cinco filas de datos y queríamos calcular el promedio de alguna columna de estos datos lo que hacíamos era leer el dato de cada fila y almacenarlo en una variable, posteriormente sumabamos estas cinco variables y la suma la dividiamos entre cinco para calcular el promedio.

Ahora imaginemos que el conjunto de datos no sólo tiene cinco filas sino miles de ellas, el procedimiento anterior sería bastante complejo, ya que almacenar cada valor en una variable tomaría demasiado tiempo con lo que esto sería impractico. Con esta observacion podemos concluir que necesitamos una mejor manera de llevar a cabo este tipo de  procesos repetitivos.

Afortunadamente Python nos ofrece una forma fácil de repetir procesos, el bucle `for`, el cual nos ayuda a repetir procesos cientos, miles o incluso millones de veces.

## 2. Bucle `for`: intro
[[Contenido]](#Contenido) 
[[Ejercicio de esta sección]](#Instrucciones-Bucle-for-intro)

Aprendamos a usar el bucle `for` para repetir un proceso varias veces. Supongamos que tenemos la lista `app_ratings = [3.5, 4.5, 4.5, 4.5, 4.5]` y el proceso que queremos llevar a cabo es: imprimir cada uno de los elementos de la lista `app_ratings`.  A continuación se muestra cómo podemos implementar este proceso utilizando el bucle `for` en Python.

```python
app_ratings = [3.5, 4.5, 4.5, 4.5, 4.5]

for element in app_ratings:
    print(element)
```

A esta técnica la llamamos **bucle** porque recorremos cada elemento de la lista y realizamos una acción con cada elemento. Debido a que siempre comenzamos con `for` (como en `for element in app_ratings:`), llamamos a esta técnica **bucle for**.

## 3. Bucle `for`: analisis
[[Contenido]](#Contenido) 
[[Ejercicio de esta sección]](#Instrucciones-Bucle-for-analisis)

En la celda de abajo se muestra el método para calcular la calificación promedio de las aplicacións (promedio de 5 aplicaciones una por cada fila).

```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
row_2 = ['Instagram', 0.0, 'USD', 2161558, 4.5]
row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]
row_4 = ['Fruit Ninja Classic', 1.99, 'USD', 698516, 4.5]
row_5 = ['Minecraft: Pocket Edition', 6.99, 'USD', 522012, 4.5]

app_data_set = [row_1, row_2, row_3, row_4, row_5]
avg_rating = (
              app_data_set[0][-1] + # Ultimo elemento de la fila 1 (row_1)
              app_data_set[1][-1] + # Ultimo elemento de la fila 2 (row_2)
              app_data_set[2][-1] + # Ultimo elemento de la fila 3 (row_3)
              app_data_set[3][-1] + # Ultimo elemento de la fila 4 (row_4)
              app_data_set[4][-1]   # Ultimo elemento de la fila 5 (row_5)
             ) / 5

print(avg_rating)
```

Afortunadamente, podemos usar un bucle `for` para calcular la calificación promedio de todas las aplicaciones. **¿Pero por donde empezamos?**. El primer paso es determinar el proceso que queremos utilizar para extraer cada calificación individual del conjunto de datos. El proceso es: **"para cada fila en** `app_data_set` **extraer el último elemento"**. En la celda de abajo se muestra cómo podemos traducir ese proceso a la sintaxis de Python e imprimir el resultado de cada iteración:

```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
row_2 = ['Instagram', 0.0, 'USD', 2161558, 4.5]
row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]
row_4 = ['Fruit Ninja Classic', 1.99, 'USD', 698516, 4.5]
row_5 = ['Minecraft: Pocket Edition', 6.99, 'USD', 522012, 4.5]

app_data_set = [row_1, row_2, row_3, row_4, row_5]

for row in app_data_set:
    rating = row[-1]
    print(rating)
```

El siguiente paso para calcular la calificación promedio de todas las aplicaciones es calcular la suma y almacenar el resultado. Aprenderemos cómo hacer esto mas adelante, pero primero es import que entendamos lo que sucede en el bucle `for` que aparece en al celda anterior. 

El bulce `for`  aísla cada fila de `app_data_set` una a la vez y la asigna a la variable `row`. Veamos un ejemplo de cómo podemos usar un bucle `for` para imprimir el contenido de cada fila del conjunto de datos `app_data_set`:

```python
app_data_set = [row_1, row_2, row_3, row_4, row_5]

for row in app_data_set:
    print(row)
```

Utilizando el bucle `for`, con solo dos líneas de código, imprimimos el contenido de las cinco listas del conjunto de datos `app_data_set`. Ahora comparemoslo con el siguiente enfoque que no utiliza el bucle `for` pero que arroja resultados equivalentes:

```python
app_data_set = [row_1, row_2, row_3, row_4, row_5]

print(app_data_set[0])
print(app_data_set[1])
print(app_data_set[2])
print(app_data_set[3])
print(app_data_set[4])
```

Usar la técnica en la celda anterior requiere que escribamos una línea de código para cada fila del conjunto de datos. Miestras que usando la técnica `for row in app_data_set` requiere que escribamos solo dos líneas de código, independientemente del número de filas en el conjunto de datos, el cual puede llegar incluso a ser de millones.

Ahora mismo uno de nuestros objetivos es utilizar la tecnica del bucle `for` para calcular la calificación promedio de las cinco aplicaciones. Un segundo objetivo es calcular la calificación promedio de un conjunto de datos mas grande. 

Un punto importante en la estructura de un ciclo `for` es el sangrado del codigo, ya que debemos **sangrar el código** que queremos que se repita cuatro caracteres de espacio a la derecha. Observa la sangría antes `rating` y `print` en la siguiente celda.

```python
app_data_set = [row_1, row_2, row_3, row_4, row_5]

for row in app_data_set:
    rating = row[-1]
    print(rating)
```

Técnicamente, solo necesitamos sangrar el código al menos un carácter de espacio a la derecha, pero la convención en la comunidad de Python es usar cuatro caracteres de espacio. Esto ayuda con la legibilidad; es decir, es más fácil para otras personas leer tu código y será más fácil para ti leer el de ellos.

## 4. Estructura del bucle `for`
[[Contenido]](#Contenido) 
[[Ejercicio 1 de esta sección]](#Instrucciones-Estructura-bucle-for-1)
[[Ejercicio 2 de esta sección]](#Instrucciones-Estructura-bucle-for-2)

Hasta ahora, hemos aprendido cómo usar bucles `for` para imprimir elementos de una lista. Ahora, veamos cómo usar bucles `for` para sumar valores de una lista. Para hacer esto, necesitamos comprender la estructura de un bucle `for`. 

```python
una_lista = [1, 2, 3]

for valor in una_lista:
    print(valor)
    print(valor * 2)
```

El código sangrado debajo del bucle `for` (el "**cuerpo**" del bucle) se ejecuta una vez para cada elemento de la lista. Entonces, si la lista tiene tres elementos, el código en el cuerpo del bucle se ejecuta tres veces. A cada ejecución del cuerpo la llamamos "**iteración**".

Durante cada iteración, la "**variable de iteración**" tomará el valor del siguiente elemento de la lista. Por ejemplo, si la lista es `[1, 3, 5]`, entonces durante la primera iteración la variable de iteración será `1`, durante la segunda iteración será `3` y durante la tercera iteración será `5`.

El código fuera del cuerpo del bucle puede interactuar con el código dentro del cuerpo del bucle. Por ejemplo, en el siguiente código, hacemos lo siguiente:

- Inicializamos una variable `a_sum` con un valor de cero fuera del cuerpo del bucle.
- Hacemos un bucle (o iteramos) sobre `a_list`. Para cada iteración del ciclo, hacemos lo siguiente:
    - Realizamos una suma (dentro del cuerpo del bucle) entre el valor actual de la variable de iteración `value` y el valor actual almacenado en `a_sum` (`a_sum` se definió fuera del cuerpo del bucle).
    - Asignamos el resultado de la suma a `a_sum` (dentro del cuerpo del bucle).
    - Imprimimos el valor de la variable `a_sum` (dentro del cuerpo del bucle). Observa que el valor de `a_sum` cambia después de cada suma. Al final del ciclo, `a_sum` tiene el valor `9`, que equivale a la suma de los números en `a_list` `(1 + 3 + 5)`.

```python
a_list = [1, 3, 5]

a_sum = 0
for value in a_list:
    a_sum = a_sum + value
    print(a_sum)

print(a_sum)
```

En la celda de anterior, creamos una forma de sumar los números en una lista. Podemos utilizar esta técnica para sumar la cantidad de calificaciones de aplicaciones en nuestro conjunto de datos.
