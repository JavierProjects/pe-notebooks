# Listas de datos
[Contenido](#Contenido)    


La siguientes celdas contienen todas las listas necesarias para generar los cinco graficos: histograma, diagrama de barras, diagrama circular, poligono de frecuencias y ojiva.  

```python
# Etiquetas del eje x
marcas_clase = ['0.165', '0.495', '0.825', '1.155', '1.485']

# Datos para el eje x
valores_ref_eje = list(range(1, len(marcas_clase) + 1))

# Freucuencia absoluta
fa = [2, 4, 4, 3, 7]

# Freucuencia relativa
fr = [0.10, 0.20, 0.20, 0.15, 0.35]

# Freucuencia acumulada
facum = [0.10, 0.30, 0.50, 0.65, 1]

# - 7 colores disponibles
# - Seleccionar color hexadecimal: https://htmlcolorcodes.com/es/
colores = ["#8A75F5", "#D419E3", "#241A58", "#19E322", "#E31972", "#198AE3", "#EECD15"]
```

<a id="Histograma"><a/>
# Histograma
[Contenido](#Contenido)    

```python
plt.figure(figsize = (8, 4)) # (Ancho, alto) del grafico

# Histograma:
# - Barras al 100% de ancho (width = 1)
# - Color del contorno de las baras en negro (edgecolor= "k")

plt.bar(valores_ref_eje, fr,
        width = 1, edgecolor= "k", 
        color = colores) 

plt.xticks(valores_ref_eje, marcas_clase, fontsize = 10) # Etiquetar marcas de clase 
plt.xlabel("Marcas de clase", fontsize = 15) # Etiqueta del eje x
plt.ylabel("Frecuencia relativa", fontsize = 15) # Etiqueta del eje y
plt.title("Histograma", fontsize = 20) # Titulo
plt.grid() # Activar cuadricula
plt.show() # Mostrar grafico
```

<a id="Barras"><a/>
# Diagrama de barras
[Contenido](#Contenido)    

```python
plt.figure(figsize = (8, 4)) # (Ancho, alto) del grafico

# Diagrama de barras:
# - Barras al 100% (width = 1)
# - Color de contorno negro (edgecolor= "k")
# - Seleccionar color hexadecimal: https://htmlcolorcodes.com/es/
plt.barh(valores_ref_eje, fa,
        height = 0.8, edgecolor= "k", 
        color = colores) 

plt.gca().invert_yaxis() # Invertir el sentido del eje y
plt.yticks(valores_ref_eje, marcas_clase, fontsize = 10) # Etiquetar marcas de clase 
plt.xlabel("Frecuencia absoluta ", fontsize = 15) # Etiqueta del eje y
plt.ylabel("Marcas de clase", fontsize = 15) # Etiqueta del eje x
plt.title("Diagrama de barras", fontsize = 20) # Titulo
plt.grid() # Activar cuadricula
plt.show() # Mostrar grafico
```

<a id="Pastel"><a/>
# Diagrama circular o de pastel
[Contenido](#Contenido)    

```python
# Validar lista separaciones NO TOCAR el codigo de abajo
if 'separaciones' in locals() or 'separaciones' in globals():
    if len(separaciones) == 0:
        print("The list exists and is empty.")
        separaciones = [0] * len(fa)

    else:
        print("The list exists and is not empty.")
else:
    print("The list doesn't exist.")
    separaciones = [0] * len(fa)
# Validar lista separaciones NO TOCAR el codigo de arriba

plt.figure(figsize = (8, 4)) # (Ancho, alto) del grafico

plt.pie(fr,
        explode = separaciones, # Editar lista 'separaciones' en caso de ser necesario
        colors = colores,
        pctdistance = .8,
        counterclock = False,
        startangle = 90,
        autopct = "%0.1f%%", # 1 valor despues del punto decimal
        labels = marcas_clase)

plt.show()
```

<a id="Poligono"><a/>
# Poligono de frecuencias
[Contenido](#Contenido)

```python
plt.figure(figsize = (8, 4)) # Ancho, alto del grafico

# Ajustes a las listas para graficar de poligono
datos_x = [0] + valores_ref_eje + [valores_ref_eje[-1] + 1]
datos_y = [0] + fr + [0]

# El poligono de frecuencias tiene: 
# - Una linea con estilo = "Linea cortadita" o discontinua. 
# - Valores validos para el estilo de la linea: '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
# - Forma de marcador triangular (marker = "v"). https://matplotlib.org/stable/api/markers_api.html

plt.plot(datos_x, datos_y, 
         linewidth = 3, color = "k" , linestyle = "--", 
         marker = "p", markersize = 10, markerfacecolor = "y", markeredgecolor = "r" ) 

plt.xticks(valores_ref_eje, marcas_clase, fontsize = 10) # Etiquetar marcas de clase 
plt.xlabel("Marcas de clase", fontsize = 15) # Etiqueta del eje x
plt.ylabel("Frecuencia relativa", fontsize = 15) # Etiqueta del eje y
plt.title("Poligono de frecuencias", fontsize = 20) # Titulo
plt.grid() # Activar cuadricula
plt.show() # Mostrar grafico
```

<a id="Ojiva"><a/>
# Ojiva
[Contenido](#Contenido)

```python
plt.figure(figsize = (8, 4)) # (Ancho, alto) del grafico

# Ajustes a las listas para graficar la ojiva
datos_x = [0] + valores_ref_eje
datos_y = [0] + facum

# El poligono de frecuencias tiene: 
# - Una linea con estilo = "Linea cortadita" o discontinua. 
# - Valores validos para el estilo de la linea: '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
# - Forma de marcador triangular (marker = "v"). https://matplotlib.org/stable/api/markers_api.html

plt.plot(datos_x, datos_y, 
         linewidth = 3, color = "k" , linestyle = "--", 
         marker = "p", markersize = 10, markerfacecolor = "y", markeredgecolor = "r" ) 

plt.xticks(valores_ref_eje, marcas_clase, fontsize = 10, ) # Etiquetar marcas de clase 
plt.xlabel("Marcas de clase", fontsize = 15) # Etiqueta del eje x
plt.ylabel("Frecuencia acumulada", fontsize = 15) # Etiqueta del eje y
plt.title("Ojiva", fontsize = 20) # Titulo
plt.grid() # Activar cuadricula
plt.show() # Mostrar grafico
```
