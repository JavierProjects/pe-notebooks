Estas son las dos reglas de sintaxis que debemos seguir cuando nombramos variables en Python:

## 4. Actualizando variables
[[Contenido]](#Contenido) 
[[Ejercicio de esta sección]](#Instrucciones-Actualizando-variables)

Podemos actualizar el valor almacenado en una variable. A continuación, primero almacenamos 1.99 en la variable `app_costs` y luego esta misma variable la actualizamos con un valor de 6.99.

```python
app_costs = 1.99
print(app_costs)

app_costs = 6.99
print(app_costs)
```

Ahora analicemos el siguiente ejemplo de actualización de variables en Python.

```python
app_costs = 1.99
app_costs = app_costs + 6.99
print(app_costs)
```

Como puedes observar en el codigo de la celda de arriba cuando ejecutamos `app_costs = app_costs + 6.99`, `app_costs` se actualiza para almacenar el resultado de `app_costs + 6.99`, que es 8.98. Ejecutar `app_costs = app_costs + 6.99` es lo mismo que ejecutar `app_costs = 1.99 + 6.99` porque `app_costs` almacena 1.99.

## 5. Atajos de sintaxis
[[Contenido]](#Contenido) 
[[Ejercicio de esta sección]](#Instrucciones-Atajos-de-sintaxis)

En el ejercicio anterior utilizamos el código `app_costs = app_costs + 6.99` para actualizar la variable `app_costs` de 1.99 a 8.98.

Hay varios atajos de sintaxis que podemos usar para actualizar una variable cuando realizamos operaciones aritméticas. Por ejemplo, como alternativa al código anterior, podemos escribir `app_costs += 6.99` (en lugar de `app_costs = app_costs + 6.99`):

```python
app_costs = 1.99
app_costs += 6.99
print(app_costs)
```

A continuación se muestra una tabla con algunos atajos de sintaxis que podemos usar:


<table>
  <tr>
    <th>Atajo de Sintaxis</th>
    <th>Ejemplo de Código</th>
    <th>Resultado</th>
  </tr>
  <tr>
    <td>+=</td>
    <td>x += 5</td>
    <td>Valor de x incrementado en 5</td>
  </tr>
  <tr>
    <td>-=</td>
    <td>y -= 3</td>
    <td>Valor de y decrementado en 3</td>
  </tr>
  <tr>
    <td>*=</td>
    <td>z *= 2</td>
    <td>Valor de z multiplicado por 2</td>
  </tr>
  <tr>
    <td>/=</td>
    <td>a /= 4</td>
    <td>Valor de a dividido por 4</td>
  </tr>
  <tr>
    <td>**=</td>
    <td>b **= 3</td>
    <td>Valor de b elevado a la potencia de 3</td>
