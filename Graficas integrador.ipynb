```python
import numpy as np
import matplotlib.pyplot as plt
print('Bibliotecas cargadas')
```

```python
fig, ax = plt.subplots(figsize=(16, 6))

mediciones = np.arange(1, 11)
altura_sin_antena = mediciones + 0.12
altura_con_antena = mediciones + 0.34

bar_width = 0.4
# Varios colores: https://matplotlib.org/stable/gallery/color/named_colors.html
b1 = ax.bar(mediciones, altura_sin_antena, width=bar_width, 
       color='greenyellow', edgecolor='black', linewidth=1, label='Sin antena')
b2 = ax.bar(mediciones + bar_width + 0.1, altura_con_antena, width=bar_width, 
       color='deepskyblue', edgecolor='black', linewidth=1, label='Antena')

# Ajustar las marcas del eje x.
ax.set_xticks(mediciones + (bar_width + 0.1) / 2)
ax.set_xticklabels(mediciones)

# Añadir textos sobre las barras
for x, ys, ya in zip(mediciones, altura_sin_antena, altura_con_antena):
    ax.text(x, ys,  f"{ys:.2f}", horizontalalignment= 'center', verticalalignment='bottom', fontsize=14)
    ax.text(x + bar_width + 0.1, ya,  f"{ya:.2f}", ha= 'center', va='bottom', fontsize=14)    

ax.legend(fontsize=20)
ax.set_ylim(0, 12)

# Añadir etiquetas y titulos
ax.set_xlabel('Mediciones', fontsize=20)
ax.set_ylabel('Altura (cm)', fontsize=20)
ax.set_title('Crecimiento del rabano', fontsize=30)

ax.tick_params(axis='x', labelsize=15)  
ax.tick_params(axis='y', labelsize=15)  
plt.grid()

# Exportar imagen
plt.savefig("Barras_transparente_500dpi.png", dpi = 500, 
           bbox_inches = 'tight', transparent = True)

plt.show()
```
