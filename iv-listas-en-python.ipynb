Una lista puede contener tipos de datos tanto mixtos como idénticos. Una lista como `[4, 5, 6]` tiene tipos de datos idénticos (solo enteros), mientras que la lista `['Facebook', 0.0, 'USD', 2974676, 3.5]` tiene datos mixtos.

La lista `['Facebook', 0.0, 'USD', 2974676, 3.5]` tiene cinco puntos de datos. Para encontrar la longitud de una lista, podemos usar la función `len()`:

```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
print(len(row_1))

list_1 = [1, 6, 0]
print(len(list_1))

list_2 = []
print(len(list_2))
```

Para listas pequeñas, podemos simplemente contar los puntos de datos en nuestras pantallas para encontrar la longitud, pero la función `len()` resultará muy útil más adelante cuando trabajemos con listas que contengan miles de elementos. ## 6. Indexación negativa
[[Contenido]](#Contenido) 
[[Ejercicio de esta sección]](#Instrucciones-Indexación-negativa)

En Python tenemos dos sistemas de indexación de listas:

**Indexación positiva**: el primer elemento tiene el número de índice 0, el segundo elemento tiene el número de índice 1, y así sucesivamente.
**Indexación negativa**: el último elemento tiene el número de índice -1, el penúltimo elemento tiene el número de índice -2, y así sucesivamente.

En la práctica, casi siempre utilizamos indexación positiva para leer elementos de una lista. La indexación negativa es útil cuando queremos leer el último elemento de una lista, especialmente si la lista es larga y no podemos saber la longitud contando.


```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]

print(row_1[-1])
print(row_1[4])
```

En la siguiente celda puedes observar que si utilizamos un número de índice que está fuera del rango de indexación, obtendremos un `IndexError`.

```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
print(row_1[6])
```

## 7. Leer multiples elementos de una lista
[[Contenido]](#Contenido) 
[[Ejercicio de esta sección]](#Instrucciones-Leer-multiples-elementos-de-una-lista)

A veces queremos leer más de un valor de una lista. Por ejemplo, si tenemos una lista como `['Facebook', 0.0, 'USD', 2974676, 3.5]`, es posible que solo queramos el nombre de la aplicación, el numero de valoraciones y la valoracion de los usuarios. Así es como podemos obtener esas partes específicas de la lista:

```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
app_name = row_1[0] # Facebook
n_of_ratings = row_1[3] # 2974676
rating = row_1[-1] # 3.5
```

Si intentamos obtener esta información para muchas aplicaciones (filas de la tabla), terminaremos con una gran cantidad de datos separados y nuestro código puede volverse largo y confuso. Una forma más sencilla de manejar esto es poner la información que queremos en una nueva lista.

```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]

fb_rating_data = [row_1[0], row_1[3], row_1[-1]]
print(fb_rating_data)
```

En el ejemplo anterior, tomamos tres elementos específicos de la lista usando el código `row_1[0], row_1[3], row_1[-1]`. Este código obtiene el primer, cuarto y último elemento de la lista. Luego colocamos estos elementos entre corchetes para crear una nueva lista.

## 8. Porciones de listas
[[Contenido]](#Contenido) 
[[Ejercicio de esta sección]](#Instrucciones-Porciones-de-listas)

En el ejercicio anterior, leimos el primer, cuarto y último elemento de una lista para obtener información especifica. Ahora, veamos cómo obtener información de datos consecutivos o contiguos de una lista. Cosidera la lista `row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]`, ahora leamos los primeros 3 valores de esta lista.

```python
row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]

cc_pricing_data = [row_3[0], row_3[1], row_3[2]]
print(cc_pricing_data)
```

El metodo de arriba es el que hemos venido usando hasta el momento pero la siguiente forma es mas facil

```python
row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]

cc_pricing_data = row_3[0:3] # Syntax shortcut
print(cc_pricing_data)
```

Este atajo es llamado "**rebanado de una lista**" (en ingles **list slicing**). Para "**rebanar**" una lista debemos hacer lo siguiente:

1. Identificar el primer y el último elemento que queremos.
2. Identificar los números de índice de esos elementos.
3. Utilizar `nombre_de_la_lista[m:n]` para obtener el segmento o rebanada, donde `m` es el índice del primer elemento y `n` es uno más que el índice del último elemento.

Considera que tenemos la lista `row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]` de la cual quremos `0.0, 'USD', 2130805`. Siguiendo los 3 pasos arriba mencionados entonces:

1. El primer elemento es  `0.0` y el ultimo es `2130805`. 
2. El indice del primer elemento `0.0` es `1` y el indice del ultimo elemento `2130805`  es `3`. 
3. Asi tenemos que `m = 1` y `n = 4` por lo que para obtener el segmento que queremos escribimos `row_3[1:4]`. A continuacion se muestra el resultado

```python
row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]

cc_pricing_data = row_3[1:4] 
print(cc_pricing_data)
```

Cuando necesitamos seleccionar los primeros o ultimos `x` elementos, podemos usar atajos de sintaxis aún más simples:

- `nombre_de_la_lista[:x]` cuando queremos seleccionar los primeros `x` elementos.
- `nombre_de_la_lista[-x:]` cuando queremos seleccionar los últimos `x` elementos.



```python
row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]

first_3 = row_3[:3] # Los primeros 3 elementos
last_3 = row_3[-3:] # Los ultimos 3 elementos

print(first_3)
print(last_3)
```

## 9. Lista de listas
[[Contenido]](#Contenido) 
[[Ejercicio de esta sección]](#Instrucciones-Lista-de-listas)

Anteriormente, se explico que las listas son una mejor alternativa que usar una variable por cada punto de datos. En lugar de tener una variable separada para cada uno de los cinco puntos de datos `'Facebook', 0.0, 'USD', 2974676, 3.5`, podemos agrupar los puntos de datos en una lista y luego almacenar la lista en una sola variable.

Hasta ahora, hemos estado trabajando con un conjunto de datos que contiene de cinco filas y hemos estado almacenando cada fila como una lista en una variable separada (las variables `row_1`, `row_2`, `row_3`, `row_4` y `row_5`). Sin embargo, si tuviéramos un conjunto de datos con 5000 filas, terminaríamos con 5000 variables donde cada una es una lista, lo que hará que sea casi imposible trabajar con nuestro código.

Para resolver este problema, podemos almacenar nuestras cinco variables (listas) en una sola lista:

```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
row_2 = ['Instagram', 0.0, 'USD', 2161558, 4.5]
row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]
row_4 = ['Fruit Ninja Classic', 1.99, 'USD', 698516, 4.5]
row_5 = ['Minecraft: Pocket Edition', 6.99, 'USD', 522012, 4.5]

data_set = [row_1, row_2, row_3, row_4, row_5]
print(data_set)
```

Observe los corchetes dobles al principio y al final del resultado que imprime la celda de arriba. Como podemos ver, `data_set` es una lista que almacena otras cinco listas (`row_1, row_2, row_3, row_4, row_5`). Una lista que contiene otras listas se llama **lista de listas**.

## 10. Leer de una lista de listas
[[Contenido]](#Contenido) 
[[Ejercicio de esta sección]](#Instrucciones-Leer-de-una-lista-de-listas)

Como se puede ver en la celda de abajo la variable `data_set` es una lista, por lo que podemos usar lo que sabemos sobre listas para obtener partes específicas de ella. Por ejemplo:

- Para obtener el primer elemento de la lista (row_1) escribimos `data_set[0]`.
- Para obtener el último elemento de la lista (row_5) escribimos `data_set[-1]`.
- Para obtener los dos primeros elementos de la lista (row_1 y row_2) escribimos `data_set[:2]`.

A continuación se muestra el código:

```python
row_1 = ['Facebook', 0.0, 'USD', 2974676, 3.5]
row_2 = ['Instagram', 0.0, 'USD', 2161558, 4.5]
row_3 = ['Clash of Clans', 0.0, 'USD', 2130805, 4.5]
row_4 = ['Fruit Ninja Classic', 1.99, 'USD', 698516, 4.5]
row_5 = ['Minecraft: Pocket Edition', 6.99, 'USD', 522012, 4.5]

data_set = [row_1, row_2, row_3, row_4, row_5]

print(data_set[0])  # Primer elemento de la lista
print(data_set[-1]) # Segundo elemento de la lista
print(data_set[:2]) # Los 2 primeros elmentos de la lista
```

A veces, es posible que queramos obtener una parte específica de una lista que está dentro de otra lista. Por ejemplo, es posible que deseemos obtener el `3.5` de `['Facebook', 0.0, 'USD', 2974676, 3.5]`, que está dentro de la lista `data_set`.

Aquí tienes una forma de hacerlo:

```python
fb_row = data_set[0] # Obtiene el primer elemento o row_1
print(fb_row) # Muestra el primer elemento o row_1
fb_rating = fb_row[-1] # Obtiene la última parte de row_1
print(fb_rating) # Muestra 3.5
```

Pero hay una forma más rápida de hacer lo mismo:

```python
print(data_set[0][-1]) # Obtiene la última parte del primer elemento o row_1
```

De ambas formas obtenemos el `3.5`, pero la segunda forma es más corta. La mayoría de la gente prefiere el camino más corto.
