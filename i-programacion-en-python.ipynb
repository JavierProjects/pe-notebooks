## 4. Sintaxis en Python
[[Contenido]](#Contenido) 
[[Ejercicio de esta sección]](#Instrucciones-Sintaxis-en-Python)

En nuestros últimos ejercicios, le dimos a la computadora dos instrucciones: `imprimir (0.00 + 6.99) e imprimir (4.5 - 3.5)`. Escribimos cada instrucción en una línea separada. Pero ¿qué pasaría si intentáramos poner ambas instrucciones en la misma línea? Vamos a intentarlo:

````python
print(0.00 + 6.99) print(4.5 - 3.5)
````

Al ejecutar la celda de arriba *recibimos un mensaje de error*:

La computadora nos dice que hay un error de sintaxis. Al igual que los lenguajes humanos, los lenguajes de programación como Python tienen reglas gramaticales llamadas sintaxis. Necesitamos seguir estas reglas cuando escribimos nuestras instrucciones, o la computadora no nos entenderá.

Por ejemplo, si hablas inglés y dices "Me gusta la ciencia de datos", la gente te entenderá. Pero si dices "me gustan los datos, la ciencia", probablemente se confundirán. Lo mismo ocurre con la programación. La computadora no entendió `print(0.00 + 6.99) print(4.5 - 3.5)` porque no sigue las reglas de sintaxis de Python.

Encontrar errores es común cuando programamos y aprenderemos constantemente sobre los errores y cómo solucionarlos.

## 7. Operaciones aritmeticas
[[Contenido]](#Contenido) 
[[Ejercicio de esta sección]](#Instrucciones-Operaciones-aritmeticas) 

Hasta el momento los programas que se han escrito solo realizan sumas y restas aunque tambien se pueden realizar operaciones como multiplicaciones, divisiones y potenciaciones en Python.

````python
print(2.3 * 2) # Multiplicación
print(6.99 / 3) # División
print(4 ** 2) # Potencia
````

El orden matemático de operaciones se refiere a las reglas que se siguen para determinar el orden en el que se deben realizar las operaciones en una expresión matemática. Ayuda a evitar ambigüedades y garantiza que todos los cálculos se realicen de manera consistente.

Hay dos acrónimos comunes utilizados para recordar el orden de las operaciones: **PEMDAS** y **BODMAS**. Ambos acrónimos representan la misma secuencia de operaciones, pero se utilizan en diferentes regiones del mundo.

1. PEMDAS: Este acrónimo es comúnmente utilizado en los Estados Unidos y significa:

   - P: Paréntesis (también conocido como paréntesis o corchetes)
   - E: Exponentes (también conocidos como potencias o raíces)
   - M: Multiplicación
   - D: División
   - A: Adición
   - S: Sustracción

Según PEMDAS, las operaciones entre paréntesis deben realizarse primero, seguidas de exponentes, multiplicación y división (de izquierda a derecha), y finalmente, adición y sustracción (de izquierda a derecha).

2. BODMAS: Este acrónimo es comúnmente utilizado en el Reino Unido y otros países, y significa:

   - B: Brackets (paréntesis)
   - O: Orden (exponentes y raíces)
   - D: División
   - M: Multiplicación
   - A: Adición
   - S: Sustracción

Bajo BODMAS, las operaciones entre paréntesis se realizan primero, seguidas de exponentes y raíces, luego división y multiplicación (de izquierda a derecha), y finalmente, adición y sustracción (de izquierda a derecha).

Comprender PEMDAS y BODMAS te permitirá construir expresiones complejas y anticipar sus resultados con precisión. Por ejemplo, conocer el orden de las operaciones es fundamental si queremos calcular el coste de comprar 10 copias de Minecraft: Pocket Edition y 10 copias de Instagram:

````python
print(6.99 + 0.00 * 10)
print((6.99 + 0.00) * 10)
````

Observando el ejemplo de código que aparece en la celda de arriba, podemos deducir de la primera operación `(6,99 + 0,00 * 10)` y su resultado correspondiente (6.99), que la multiplicación precede a la suma. ¡Esta operación no incluye el costo de comprar 10 copias de `Minecraft: Pocket Edition`! por lo que es incorrecta.

Sin embargo, en la segunda operación ((6,99 + 0,00) * 10), la suma se calcula primero porque esta vez está entre paréntesis. En consecuencia, el resultado es 69.9, el cual es correcto.

Hasta ahora, hemos utilizado caracteres de espacio entre números y operadores (`+, -, *, /, **` son operadores). Por ejemplo, hemos utilizado `6.99 + 0.00` en lugar de `6.99+0.00`. Tanto `6.99 + 0.00` como `6.99+0.00` se ejecutarán correctamente en Python; sin embargo, te recomendamos que utilices espacios en tu código porque mejora la legibilidad.
